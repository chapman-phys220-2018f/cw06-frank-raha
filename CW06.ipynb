```python
import elementary as elm
import numpy as np

# Basic plotting defaults
import matplotlib.pyplot as plt
%matplotlib inline

# Use better color defaults
plt.style.use("seaborn-pastel")
```

```python
#Creating particles at respective coordinates
particle = elm.Particle(0,0,1)
electron = elm.Electron(0,0,2)
proton = elm.Proton(0,0,3)
```

```python
#Each particle has a different mass
print(particle.mass)
print(electron.mass)
print(proton.mass)
```

```python
#Giving momentum of 10 kg m/s
particle.impulse(0,0,10*particle.mass)
electron.impulse(0,0,10*electron.mass)
proton.impulse(0,0,10*proton.mass)

#Setting figure size and axis
f = plt.figure(figsize=(8,6))
a = plt.axes()

x_points = [i for i in np.arange(0,5,1e-2)]

#Empty list of particle y positions
y_particle = []
y_electron = []
y_proton = []

#Finding points for each particle as it is affected by gravity
for i in range(500):
    particle.impulse(0,0,-9.8*particle.mass*.01)
    particle.move(1e-2)
    y_particle.append(particle.position[2])

for i in range(500):
    electron.impulse(0,0,-9.8*electron.mass*.01)
    electron.move(1e-2)
    y_electron.append(electron.position[2])

for i in range(500):
    proton.impulse(0,0,-9.8*proton.mass*.01)
    proton.move(1e-2)
    y_proton.append(proton.position[2])

#Plot each particle with corresponding x and y points
a.plot(x_points, y_particle, color="Red", label="Partice")
a.plot(x_points, y_electron, color="Blue", label="Partice")
a.plot(x_points, y_proton, color="Green", label="Partice")

#Show graph with labeled axes
a.set(xlabel = "Time (s)", ylabel = "Position (m)", title = "Particle Trajectory")
a.legend
plt.show()
```

#### ANALYSIS
Despite each mass being different, the trajectories based on gravity are the same; each trajectory is very similar.
